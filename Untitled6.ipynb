```python

import gradio as gr
from TTS.api import TTS

# Load a WaveNet-style TTS model (Tacotron2 + DDC + vocoder)
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)

def generate_audio(text):
    # Generate waveform from text
    audio = tts.tts(text)
    sample_rate = tts.output_sample_rate
    return (sample_rate, audio)

# Create a simple Gradio UI
interface = gr.Interface(
    fn=generate_audio,
    inputs=gr.Textbox(lines=2, placeholder="Type something to speak..."),
    outputs=gr.Audio(label="Generated Audio"),
    title="WaveNet-style Text-to-Speech"
)

# Launch the app
interface.launch()
```
